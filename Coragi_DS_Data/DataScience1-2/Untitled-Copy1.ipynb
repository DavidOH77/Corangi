```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3],[90,100,50])
plt.show()
```

```python
plt.plot(['Math', 'Korean', 'English'], [90, 100, 50])
plt.xlabel('Subjects')
plt.ylabel('Score')
plt.title('Midterm')

plt.plot(['Math', 'Korean', 'English'], [70, 60, 90])
plt.legend(['OH', 'KIM'])

plt.show()
```

```python
# 막대 차트 그리기
x = ['Math', 'Korean', 'English']
y = [90, 100, 50]

plt.xlabel('Subjects')
plt.ylabel('Score')

plt.bar(x, y, width = 0.7, color = 'pink')
plt.show()
```

```python
plt.plot([1,2,3], [90, 100, 50], marker='o', markersize = 100, linestyle = ':')

plt.grid(True)
plt.axis([0, 5, 50, 90])

plt.tight_layout()

plt.show
```

```python
import pandas as pd
train = pd.read_csv('data/train.csv')

train.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
sns.relplot(data=train, x = 'Age', y = 'Survived')
```

```python
sns.relplot(data=train, x ='Age', y ='Survived', hue = 'Sex', aspect = 2)
```

```python
#Categorical Scatter Plot

sns.catplot(data=train, x = 'Pclass', y= 'Fare', aspect =4)

sns.catplot(data=train, x = 'Pclass', y= 'Fare', jitter = False, aspect =4)

sns.catplot(data=train, x = 'Pclass', y= 'Fare', aspect =4, hue='Sex')
```
