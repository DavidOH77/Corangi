```python
# 라이브러리 가져오기

import requests
from bs4 import BeautifulSoup
```

```python
# 사이트 접근하기

raw = requests.get("https://search.naver.com/search.naver?where=news&query=호랑이")
```

```python
# 파싱하기

html = BeautifulSoup(raw.text, "html.parser")
```

```python
# 컨테이너 선택하기
# 컨테이너는 ul.type01 > li 이다.

articles = html.select("ul.type01 > li")
```

```python
# 제목을 나타내는 선택자 선택하고 텍스트로 변환하기
# 제목을 나타내는 선택자는 a._sp_each_title 이다.

title = articles[0].select_one("a._sp_each_title").text
```

```python
# 제목 출력하기

print(title)
```

```python
for article in articles:
    title = article.select_one("a._sp_each_title").text
    source = article.select_one('span._sp_each_source').text
    
    print('제목 : ' + title,'|','출처 : ' + source)
    print('='*60)
```

```python
# 여러 페이지 접근하기

MAX = 100

for n in range (1, MAX, 10) : #1부터 MAX까지 10씩 증가시키기
    raw = requests.get("https://search.naver.com/search.naver?where=news&query=호랑이&start=" + str(n))
    
    html = BeautifulSoup(raw.text, 'html.parser')
    
    articles = html.select('ul.type01 > li')
    
    for article in articles:
        title = article.select_one("a._sp_each_title").text
        source = article.select_one('span._sp_each_source').text
    
        print('제목 : ' + title,'|','출처 : ' + source)
        print('='*60)
```
