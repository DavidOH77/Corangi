```python
# 라이브러리 가져오기
import requests
from bs4 import BeautifulSoup
```

```python
# 사이트 접근하기
raw = requests.get("https://tv.naver.com/r")
```

```python
print(raw)
```

```python
print(raw.text)
```

```python
print(raw.elapsed)
```

```python
# 파싱하기 (무의미한 코드 > 유의미한 코드)
html = BeautifulSoup(raw.text, "html.parser")
```

```python
print(html)
```

```python
# 컨테이너 선택하기
# .select('선택자') 는 선택자를 사용해서 선택자에 해당하는
# 데이터를 모두 가져온다.
# 이렇게 가져온 데이터는 리스트 형식으로 변수에 저장된다.

clips = html.select("div.inner")
```

```python
print(clips[0])
```

```python
# .select_one("선택자") 는 선택자를 사용해서 선택자에 해당하는
# 데이터 중에서 첫번째 데이터를 가져온다.

title = clips[0].select_one("dt.title")
```

```python
print(title)
```

```python
# .text 는 선택한 데이터를 텍스트 형식으로 바꿔준다.

print(title.text)
```

```python
# .strip() 은 앞뒤의 공백을 없에준다.

print(title.text.strip())
```
