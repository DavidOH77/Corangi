```python
import requests
from bs4 import BeautifulSoup
```

```python
f= open('naver_tv.csv', 'w')
f.write('제목, 채널 명, 재생 수, 좋아요 수\n') # 미리 입력할 데이터

raw = requests.get("https://tv.naver.com/r")

html = BeautifulSoup(raw.text, "html.parser")

clips = html.select(".cds_info")

n = 0

for clip in clips:
    title = clip.select_one('dt.title').text.strip()
    chn = clip.select_one('dd.chn').text.strip()
    hit = clip.select_one('span.hit').text.strip()
    like = clip.select_one('span.like').text.strip()
    
    title = title.replace(',', '')
    title = title.replace('※','')
    title = title.replace('♬','')
    title = title.replace('♥','')
    title = title.replace('↑','')
    title = title.replace('↗','')
    title = title.replace('♡','')
    title = title.replace('♥','')
    title = title.replace('♪','')
    title = title.replace('♨','')
    title = title.replace('♫','')
    title = title.replace('＞','')
    title = title.replace('▿','')
    title = title.replace('＜','')
    title = title.replace('♭','')
    title = title.replace('★','')
    title = title.replace('☆','')
    title = title.replace('↘','')
    title = title.replace('◉','')
    title = title.replace('0','')
    title = title.replace('◉','')
    
    title = title.replace('♩','')
    title = title.replace('♯','')
    title = title.replace('–','')

    title = title.replace('♮', '')
    
    
    chb = chn.replace(',', '')
    hit = hit.replace(',', '')
    like = like.replace(',', '')
    
    hit = hit[4:]
    like = like[5:]
    
    
    
   # _^-^* ) ( ] [ + ?
    print(title)
    print('\n')
    print(chn)
    print('\n')
    print(hit.replace('수','수 : '))
    print('\n')
    print(like.replace('수','수 : '))
    print('='* 60)
    n = n + 1
    print(n)
    f.write(title + ',' + chn + ',' + hit + ',' + like + '\n')
    
    
    
f.close()
```

```python
# 파일을 엑셀로 저장합니다.

import openpyxl
```

```python
# 새 워크북을 만듭니다.
wb = openpyxl.Workbook()

# 해당하는 워크북을 원하는 이름으로 저장합니다.
wb.save('test.xlsx')
```

```python
wb = openpyxl.Workbook()

sheet = wb.active # 현재활성화된 시트를 선택합니다.
sheet['A1'] = 'hello world' # 셀번호를 통해 셀 값을 입력하거나 변경합니다.
sheet.cell(row=3, column=3).value = 'BYE!!' # 행/열 번호를 통해 셀 값을 입력하거나 변경합니다.

subject = ['Python', 'Java', 'HTML', 'JavaScript']
sheet.append(subject) #시트에 행을 추가합니다. 이때 위와 같은 리스트의 형태로 추가해야 합니다.

sheet1 = wb.active
sheet1.title = '1st sheet'

sheet2 = wb.create_sheet('2nd sheet')

for i in range(1, 10):
    sheet1.cell(row = 1, column = 1).value = i
    sheet2.cell(row = 1, column = i).value = i


wb.save('test2.xlsx')
```
