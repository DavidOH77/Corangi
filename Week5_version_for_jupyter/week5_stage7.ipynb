<h1>'함께실습$1' <h1>Iris데이터 불러오기

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data_2/iris.csv')
df.head(10)
```

```python
df.info()
```

<h1>'함께실습$2' <h1>꽃받침 길이, 너비에 따른 종 분포도 보기

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()
```

```python
sns.catplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', aspect=2)
```

<h1>'미션$1' <h1>꽃잎의 길이, 너비에 따른 종 분포도 보기

```python
sns.catplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species', aspect=2)
```

<h1>'함께실습$3' <h1>kdeplot으로 꽃받침 길이별 밀집 수준 파악

```python
facet = sns.FacetGrid(df, hue="Species", aspect=2)
facet.map(sns.kdeplot, 'SepalLengthCm')
facet.add_legend()

plt.show()
```

<h1>'함께실습$4' <h1>box plot으로 종별 꽃받침 길이 범위 파악

```python
sns.catplot(kind='box', data=df, x='Species', y='SepalLengthCm', aspect=2)
```

<h1>'자율$1' <h1>violin plot - x: 종, y: 꽃받침 길이

```python
sns.violinplot(data=df, x='Species', y='SepalLengthCm')
```

<h1>'자율$2' <h1>violin plot - x: 종, y: 꽃받침 너비

```python
sns.violinplot(data=df, x='Species', y='SepalWidthCm')
```

<h1>'자율$3' <h1>violin plot - x: 종, y: 꽃잎 길이

```python
sns.violinplot(data=df, x='Species', y='PetalLengthCm')
```

<h1>'자율$4' <h1>violin plot - x: 종, y: 꽃잎 너비

```python
sns.violinplot(data=df, x='Species', y='PetalWidthCm')
```
