<h1>'함께실습$1' <h1>Count Plot으로 Age에 따른 생존여부 히스토그램 얻기

```python
# Count Plot으로 Age에 따른 생존여부 히스토그램 얻기

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
sns.set()
```

```python
sns.catplot(data = df, x = 'Age' , hue = 'Survived', kind = 'count', aspect = 4)
```

<h1>'함께실습$2' <h1>kdeplot을 사용하여 매끄러운 곡선 분포도 얻기

```python
# kdeplot을 사용하여 매끄러운 곡선 분포도 얻기

facet = sns.FacetGrid(df, hue = 'Survived', aspect = 4)
facet.map(sns.kdeplot, 'Age')
facet.add_legend()

plt.show()

# y축은 seaborn이 kernel density를 측정하여 밀집 정도를 계산한 값
# 각 값이 무엇인지 중요한게 아니라, 높을수록 밀집, 낮을수록 희소 하다는 의미만 기억하면 된다.

```
